### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target

np.random.seed(5)
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
print(Ｘ.shape)
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import decomposition

# demo範例沒有先做標準化, 自己建議先做
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
X_std = StandardScaler().fit_transform(X)

centers = [[1, 1], [-1, -1], [1, -1]]
#PCA 先取降為3維 畫立體圖
X_pca = decomposition.PCA(n_components=3).fit_transform(X_std)
print(X_pca.shape)
X_pca
```

```python
#先取標準化在PCA , 畫立體圖觀察

fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()


for name, label in [('y=0', 0), ('y=1', 1), ('y=2', 2)]:
    ax.text3D(X_pca[y == label, 0].mean(),
              X_pca[y == label, 1].mean() + 1.5,
              X_pca[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
#y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
#PCA 先取降為3維 畫立體圖
X_pca2 = decomposition.PCA(n_components=3).fit_transform(X)
print(X_pca2.shape)
X_pca2
```

```python
#先取標準化在PCA , 畫立體圖觀察

fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()


for name, label in [('y=0', 0), ('y=1', 1), ('y=2', 2)]:
    ax.text3D(X_pca2[y == label, 0].mean(),
              X_pca2[y == label, 1].mean() + 1.5,
              X_pca2[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
#y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X_pca2[:, 0], X_pca2[:, 1], X_pca2[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
